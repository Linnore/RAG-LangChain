```python
import pandas as pd 
import numpy as np

df = pd.read_csv("raw/data_label.csv")
output_path = "./docs/QA/QA_reference_{range}.txt"
size = 1000
lines = []

np.random.seed(118010142)
rand_idx = np.arange(df.shape[0])
np.random.shuffle(rand_idx)
for idx in rand_idx:
    row = df.iloc[idx]
    line = "问题：{}{} 回答：{}\n\n".format(row["title"], row["ask"], row["answer"])
    if (idx+1)%size==0:
        with open(output_path.replace("{range}", f"{str(idx-size+1)}-{str(idx)}"), "w", encoding="utf-8") as outfile:
            outfile.writelines(lines)
            lines = []
    else:
        lines.append(line)
        
with open(output_path.replace("{range}", f"{str(idx-((idx+1)%size)+1)}-{str(idx)}"), "w", encoding="utf-8") as outfile:
    outfile.writelines(lines)
```
